```python
from math import sqrt
from sklearn.metrics import mean_squared_error
X = data
size_to_divide = int(len(X) * 0.66)
train, test = X[0:size_to_divide], X[size_to_divide:len(X)]
history_data = [x for x in train]
y_predictions = list()
# walk-forward validation
for t in range(len(test)):
	model_arima = ARIMA(history_data, order=(5,1,0))
	model_arima_fit = model_arima.fit()
	output = model_arima_fit.forecast()
	y_pred = output[0]
	y_predictions.append(y_pred)
	obs = test[t]
	history_data.append(obs)
	print('predicted by model = ',y_pred,'expected= ', obs)
 
# calculating root mean square error
root_mean_square = sqrt(mean_squared_error(test, y_predictions))

print('Test root mean square = ', root_mean_square)

# plot forecasts against actual outcomes
pyplot.plot(test)
pyplot.plot(y_predictions, color='green')
pyplot.show()
```

```python

# converting to supervised problem
def to_supervised(data, lag_factor=1):
	df_temp = pd.DataFrame(data)
	column = [df_temp.shift(i) for i in range(1, lag_factor+1)]
	column.append(df_temp)
	df_temp = pd.concat(column, axis=1)
	df_temp.fillna(0, inplace=True)
	return df_temp
```

```python
X = data
supervised_data = to_supervised(X, 1)
print(supervised_data.head())
```

```python

# calculating difference between the rows
def difference_between(dataset, interval=1):
	diff = list()
	for i in range(interval, len(dataset)):
		value = dataset[i] - dataset[i - interval]
		diff.append(value)
	return pd.Series(diff)



# invert the values
def inverse_difference_between(history_data, ypredicted, interval=1):
	return ypredicted + history_data[-interval]


difference_values = difference_between(data, 1)
print("difference between rows at 1 interval : \n",difference_values.head())

# invert the difference calculated
inverted_difference = list()
for i in range(len(difference_values)):
	value = inverse_difference_between(data, difference_values[i], len(data)-i)
	inverted_difference.append(value)
inverted_difference = pd.Series(inverted_difference)
print("inveretd difference between rows at interval 1 : \n ", inverted_difference.head())
```

```python

from sklearn.preprocessing import MinMaxScaler

 
# scaling the data between -1 to 1
def scale_data(train, test):
	# fit min max scaler
	minmax_scaler = MinMaxScaler(feature_range=(-1, 1))
	minmax_scaler = minmax_scaler.fit(train)
	# train
	train = train.reshape(train.shape[0], train.shape[1])
	train_scaled = minmax_scaler.transform(train)
	# test
	test = test.reshape(test.shape[0], test.shape[1])
	test_scaled = minmax_scaler.transform(test)
	return minmax_scaler, train_scaled, test_scaled
 

# inverse scaling
def invert_scale_data(scaler, X, value):
	temp_data = [x for x in X] + [value]
	array_data = np.array(temp_data)
	array_data = array_data.reshape(1, len(array_data))
	data_inverted = scaler.inverse_transform(array_data)
	return data_inverted[0, -1]
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM

# LSTM Firt
def fit_lstm(train_data, batchSize, epoch, neurons):
	X, y = train_data[:, 0:-1], train_data[:, -1]
	X = X.reshape(X.shape[0], 1, X.shape[1])
	model_lsstm = Sequential()
	model_lsstm.add(LSTM(neurons, batch_input_shape=(batchSize, X.shape[1], X.shape[2]), stateful=True))
	model_lsstm.add(Dense(1))
 
	model_lsstm.compile(loss='mean_squared_error', optimizer='adam')
	for i in range(epoch):
		model_lsstm.fit(X, y, epochs=epoch, batch_size=batchSize, verbose=0, shuffle=False)
		model_lsstm.reset_states()
	return model_lsstm
 
# make forecast
def forecast_lstm(model_lstm, batch_size, X):
	X = X.reshape(1, 1, len(X))
	y_pred = model_lstm.predict(X, batch_size=batch_size)
	return y_pred[0,0]
 
```

```python
# convert to stationary data
original_values = data
difference_values = difference_between(original_values, 1)
 
# convert to supervised data
supervised = to_supervised(difference_values, 1)
supervised_data = supervised.values
 
# split data into train_data and test-sets
train_data, test_data = supervised_data[0:-15], supervised_data[-15:]
 
# transform the scale of the data
minmax_scaler, train_data_scaled, test_data_scaled = scale_data(train_data, test_data)
 
# fitting the model
lstm_model_ = fit_lstm(train_data_scaled, 1, 10, 4)

# prediction on training data
train_data_reshaped = train_data_scaled[:, 0].reshape(len(train_data_scaled), 1, 1)
lstm_model_.predict(train_data_reshaped, batch_size=1)
 
# walk-forward validation for test data
y_predictions = list()
for i in range(len(test_data_scaled)):

	X, y = test_data_scaled[i, 0:-1], test_data_scaled[i, -1]
	y_pred = forecast_lstm(lstm_model_, 1, X)
	y_pred = invert_scale_data(minmax_scaler, X, y_pred)
	y_pred = inverse_difference_between(original_values, y_pred, len(test_data_scaled)+1-i) 
	y_predictions.append(y_pred)
	expected = original_values[len(train_data) + i + 1]
	print('Day=%d, Predicted=%f, Expected=%f' % (i+1, y_pred, expected))
 
# calculating evalution metrics
root_mean_square_metric = sqrt(mean_squared_error(original_values[-15:], y_predictions))
print('test data root_mean_square_metric: %.3f' % root_mean_square_metric)

# plot
pyplot.plot(original_values[-15:])
pyplot.plot(y_predictions)
pyplot.show()
```
